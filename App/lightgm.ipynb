```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
y_train = train['count']
X_train = train.drop(['count', 'atemp'], axis=1)
y_test = test['count']
X_test = test.drop(['count', 'atemp'], axis=1)
```

```python
X_train = np.log1p(X_train)
X_test = np.log1p(X_test)
```

```python
X_train
```

```python
from lightgbm import LGBMRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_squared_log_error
```

```python
lgb = LGBMRegressor(learning_rate=0.05, n_estimators=100, num_leaves=1000, max_depth=-1, min_child_samples=60)
```

```python
lgb.fit(X_train, y_train)
pred = lgb.predict(X_test)

train_score = lgb.score(X_train, y_train)
r2 = r2_score(y_test, pred)
mae = mean_absolute_error(y_test, pred)
mse = mean_squared_error(y_test, pred, squared=True)
rmse = mean_squared_error(y_test, pred, squared=False)

print(f'Train r2 score: ', train_score, '\n'
      'Test: ', '\n'
      ' - r2 score: ',  r2, '\n'
      ' - mae: ',  mae, '\n'
      ' - mse: ',  mse, '\n'
      ' - rmse: ',  rmse, '\n'
      )
```
